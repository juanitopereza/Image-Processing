```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img=np.array(mpimg.imread("lambda_21.png",format="rbg"))
imgplot = plt.imshow(img)
plt.show()

```

```python
img_filter = np.copy(img)
img_filter[img_filter < 20] = 0
img_filter[img_filter > 200] = 0
```

```python
plt.imshow(img_filter)
```

```python
grid_x, grid_y = np.meshgrid(range(np.shape(img)[0]), range(np.shape(img)[1]))
```

```python
region_1 = grid_y < 800
region_2 = 420 < grid_x
region_3 = grid_x < 750
region_4 = region_1*region_2*region_3
```

```python
plt.imshow(region_4)
```

```python
img_cuadro = np.copy(img)
img_cuadro[:,:,0] *= region_4.T
img_cuadro[:,:,1] *= region_4.T
img_cuadro[:,:,2] *= region_4.T
```

```python
plt.imshow(img_cuadro)
plt.scatter(730,580)
```

```python
circulei = (grid_x - 580)**2 + (grid_y - 730)**2 < 100**2
```

```python
carita = np.copy(img)
carita[:,:,0] *= circulei.T
carita[:,:,1] *= circulei.T
carita[:,:,2] *= circulei.T
```

```python
plt.imshow(carita)
```

```python
def tristeza(img):
    return 0.299*img[:,:,0] + 0.587*img[:,:,1] + 0.114*img[:,:,2]
```

```python
perrito_triste = tristeza(img)
```

```python
plt.imshow(perrito_triste, cmap='gray')
```

```python
perrito_triste_carita = perrito_triste*circulei.T
```

```python
plt.imshow(perrito_triste_carita, cmap='gray')
```
